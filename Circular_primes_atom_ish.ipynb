```python
# Atom-ish visualization Circular Primes
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict

xp = [2, 3, 5, 7, 11, 13, 17, 31, 37, 71, 73, 79, 97, 113, 131, 197, 199, 311, 337, 373, 719, 733, 919, 971, 991, 1193, 1931, 3119, 3779, 7793, 7937, 9311, 9377, 11939, 19391, 19937, 37199, 39119, 71993, 91193, 93719, 93911, 99371, 193939, 199933, 319993, 331999, 391939, 393919, 919393, 933199, 939193, 939391, 993319, 999331, 1111111111111111111, 11111111111111111111111]
# List for the ammount of digits in the numbers
digits = [len(str(x)) for x in xp]

# Makes a group for each digit length
groups = defaultdict(list)
for num, d in zip(xp, digits):
    groups[d].append(num)

# Makes a circular plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Puts the groups in 'orbitals' based on digit length ands spaces evenly
for d, nums in groups.items():
    n_points = len(nums)
    angles = np.linspace(0, 2 * np.pi, n_points, endpoint=False)
    radii = np.full(n_points, d)
    ax.scatter(angles, radii, label=f'{d} digits')
# Color of all dots can be changed in ax.scatter
# Titles, legends, colors text etc
ax.legend(title='Number of Digits',fontsize= 13, title_fontsize= 13)
# Hides the standard degrees of the radial plot
ax.set_xticks([])
ax.set_yticklabels([])
# Background color and text
fig.patch.set_facecolor('#000930')
ax.annotate('999331:', xy=(0.04, 6.5), color= 'brown',fontsize= 11)
ax.annotate('Last non-repunit C.P.', xy=(-0.2, 6.5), color= 'brown',fontsize= 11)
ax.annotate('$R$19', xy=(0.04, 17.5), color= 'magenta',fontsize= 11)
ax.annotate('$R$23', xy=(0.04, 21.5), color= 'grey',fontsize= 11)
ax.annotate('0', xy=(0, 0), ha='center', va='center', fontsize=11, color='#c71585')
# Sets the max circles (the maximum + 1)
ax.set_yticks(range(1, max(digits) + 1))

# Removes all 'orbitals' without values
filledorbitals = list(groups.keys())
ax.set_yticks(filledorbitals)

plt.show()
```
