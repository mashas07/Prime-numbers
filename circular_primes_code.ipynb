```python
# Code for generating Circular Primes
# The rotator (makes the cycle of a number)
def rotator(i):
    n = str(i)
    nrot = n[1:] + n[0]
    rotations= []
    rotations.append(n)
    while n != nrot:
        rotations.append(nrot)
        nrot= nrot[1:] + nrot[0]
    irotations = [int(x) for x in rotations]
    return(irotations)

print(rotator(197))

# Sieve of Erastosthenes (generates the primes)
def sieve(n):
    if n <= 2:
        return []
    isprime = [True] * n
    isprime[0] = False
    isprime[1] = False

    for i in  range (2, int(n**0.5+1)+1):
            for x in range(i*i, n, i):
                isprime[x] = False
    return [i for i in range(n) if isprime[i]]

print((sieve(100)))

# The list generator (checks whether ALL the generated rotations for a specific p are in the list of primes)
def cprimes(c):
    primes = set(sieve(c))
    cprimelist = []

    for p in primes:
        rots = rotator(p)
        if all(r in primes for r in rots):
            cprimelist.append(p)

    return sorted(cprimelist)

print(cprimes(100000))
#https://www.worldofnumbers.com/circular.htm#goditable
```
